```python
x = []
y = []
t = np.linspace(0, 2, 100)
x = np.cos(t * np.pi)
y = np.sin(t * np.pi)
```

```python
plt.plot(x, y)
```

```python
W = np.random.randn(2,2)
```

```python
O = np.row_stack([x, y])
T = np.linalg.inv(W).dot(O)
```

```python
WT = W.dot(T)
plt.plot(WT[0], WT[1])
```

```python
x = np.sort(np.random.randn(10))
y = np.sort(np.random.randn(10))
```

```python
_x = x.copy()
_y = y.copy()
```

```python
X = x.reshape((10, 1))
```

```python
w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
plt.scatter(x, y, s=10)
plt.plot(x, x * w)

alpha = 5
w2 = np.linalg.inv(X.T.dot(X) + alpha * np.eye(X.T.dot(X).shape[0])).dot(X.T).dot(y)
plt.plot(x, x * w2)
```
