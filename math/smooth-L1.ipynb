$$
smoothL_1(x) = \begin{cases} 
0.5x^2 & |x|<1, \\
|x|-0.5 & otherwise
\end{cases}
$$

$$
smoothL_1(x) = \begin{cases} 
-1 & x \le -1, \\
x & |x|<1, \\
1 & x \ge 1
\end{cases}
$$

```python
x = np.linspace(-5,5,100)
```

```python
y = np.zeros_like(x)
dy = np.zeros_like(x)
```

```python
y[np.abs(x) < 1] = 0.5 * x[np.abs(x) < 1] ** 2
dy[np.abs(x) < 1] = x[np.abs(x) < 1]
```

```python
y[np.abs(x) >= 1] = np.abs(x[np.abs(x) >= 1]) - 0.5
dy[x >= 1] = 1
dy[x <= -1] = -1
```

```python
plt.plot(x, y, label="$smooth-L_1$")
plt.plot(x, dy, label="$d[{smooth-L_1}]/dx$")
plt.legend()
```
